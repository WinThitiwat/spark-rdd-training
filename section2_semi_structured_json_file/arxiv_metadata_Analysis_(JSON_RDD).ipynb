```python
# Initializing Spark
from pyspark import SparkContext, SparkConf

conf = SparkConf() \
        .setAppName("Archive_PySpark") \
        .setMaster("local[*]")

sc = SparkContext(conf=conf).getOrCreate()

print(sc)
print('ready to go')
```

```python
# Read and Load Data to Spark
# Data source: https://www.kaggle.com/Cornell-University/arxiv/version/62

import json

rdd_json = sc.textFile("data/arxiv-metadata-oai-snapshot.json", 100)

rdd = rdd_json.map(lambda json_line: json.loads(json_line))

rdd.persist() # to avoid calling the transformation function repeatedly

print("data persisted")
```

```python
# Check the number of parallelism and partitions:
print(sc.defaultParallelism)
print(rdd.getNumPartitions())
```

## Question 6: Find abbreviations with 5 or more letters in the abstract

```python
import re

def get_abbreviation(line):
    result = re.search(r"\(([a-zA-Z][^_ /\\<>]{5,})\)", line)
    if result:
        return result.group(1)
```

```python
rdd.filter(lambda x: get_abbreviation(x["abstract"])).count()
```

```python
rdd.filter(lambda x: get_abbreviation(x["abstract"])).take(5)
```

## Question 7: Get the number of archive records per month ('update_date' attribute)

```python
import datetime

def extract_date(dateln, datetype="month"):
    d = datetime.datetime.strptime(dateln, "%Y-%m-%d")
    if datetype=="day":
        print(d.day)
        return d.day
    elif datetype=="year":
        return d.year
    else:
        return d.month
```

```python
rdd.map(lambda x: (extract_date(x["update_date"]), 1)).reduceByKey(lambda x, y: x+y).collect()
```

## Question 8: Get the average number of pages

```python
import re

def get_page(line):
    search_found = re.findall("\d+ pages", line)
    if search_found:
        return int(search_found[0].split(" ")[0])
    else:
        return 0

# get_page("60 pages")
```

```python
rdd_avg = rdd.map(lambda x: get_page(x["comments"] if x["comments"] != None else "None"))

rdd_avg = rdd_avg.filter(lambda x: x != 0)

avg_counter = rdd_avg.count()
avg_sum = rdd_avg.reduce(lambda x,y: x+y)

# print(f"avg_counter: {avg_counter}")
# print(f"avg_sum: {avg_sum}")
print("The average of pages is: ", (avg_sum/avg_counter))
```
